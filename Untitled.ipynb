```python
def extract_model_info(model_name):
    gender = None
    product_type = None
    color = None

    if pd.isna(model_name):
        return None, None, None

    model_name_str = str(model_name)

    # Extract gender
    if '남성' in model_name_str:
        gender = '남성'
    elif '여성' in model_name_str:
        gender = '여성'
    elif 'UNI' in model_name_str:
        gender = '남여공용'

    # Extract color from parentheses
    color_match = re.search(r'\((.*?)\)', model_name_str)
    if color_match:
        color = color_match.group(1).strip()
        # Clean color to remove codes like 'Z1' or 'K7'
        color = re.sub(r'\s*\([A-Za-z0-9]+\)', '', color).strip()
        if color == '':
            color = None
        else:
            color = re.sub(r'\s*(Z1|K7|W1|Y1|G1|C1|B1|D1|E1|F1|H1)\s*', '', color).strip()
            if color == '':
                color = None

    # Extract product type
    product_type = model_name_str
    if gender:
        product_type = product_type.replace(gender, '').strip()
    if color_match:
        product_type = product_type.replace(color_match.group(0), '').strip()
    product_type = re.sub(r'[\(\[\{\}\]\)]', '', product_type).strip() # Remove any remaining parentheses or brackets

    # Remove the gender from the product type (again, to be safe)
    product_type = product_type.replace('남성', '').replace('여성', '').replace('UNI', '').strip()

    # Remove any leading/trailing special characters
    product_type = re.sub(r'^[-\s,]+|[-,,\s]+$', '', product_type).strip()

    return product_type if product_type else None, gender, color
```

```python
# Apply the function to create new columns
df[['Product_Type', 'Gender', 'Extracted_Color']] = df['Model_Name'].apply(
    lambda x: pd.Series(extract_model_info(x))
)
```

```python
# Clean and standardize `Color` column
df['Color'] = df['Color'].replace('', np.nan)
df['Color'] = df['Color'].fillna(df['Extracted_Color'])

def clean_color(color_name):
    if pd.isna(color_name):
        return None
    cleaned_color = re.sub(r'\s*\(.*\)\s*', '', str(color_name)).strip()
    return cleaned_color if cleaned_color else None

df['Color'] = df['Color'].apply(clean_color)

# Drop the temporary `Extracted_Color` column
df.drop(columns=['Extracted_Color'], inplace=True)
```

```python
# --- 4. Combine and standardize material information ---
df['Product_Material'] = df['Product_Material'].fillna('')
df['Main_Product_Material'] = df['Main_Product_Material'].fillna('')
df['Material'] = df['Material'].fillna('')

df['Combined_Product_Material'] = df['Product_Material'] + ' ' + df['Main_Product_Material'] + ' ' + df['Material']
```

```python
def remove_duplicate_materials(text):
    parts = re.split(r'[/,]', text)
    cleaned_parts = []
    seen = set()
    for part in parts:
        part = part.strip()
        if part not in seen:
            cleaned_parts.append(part)
            seen.add(part)
    return ', '.join(cleaned_parts)

df['Combined_Product_Material'] = df['Combined_Product_Material'].apply(remove_duplicate_materials)

common_materials = [
    '면', '아크릴', '나일론', '폴리에스터', '폴리우레탄', '솜털', '깃털',
    '우모', '거위', '오리', '알루미늄', '천연가죽', '합성피혁', 'I.P', 'PHYLON',
    'PU', 'PVC', '두랄루민', 'Carbon', 'Carbonal', '신세틱', '러버'
]

material_patterns = {
    '면': r'(?:면|COTTON|코튼)\s*(\d+)%',
    '아크릴': r'(?:아크릴|ACRYLIC)\s*(\d+)%',
    '나일론': r'(?:나일론|NYLON)\s*(\d+)%',
    '폴리에스터': r'(?:폴리에스터|POLYESTER|POLYESER|폴리에스텔)\s*(\d+)%',
    '폴리우레탄': r'(?:폴리우레탄|POLYURETHANE|SPANDEX)\s*(\d+)%',
    '솜털': r'(?:솜털|DOWN)\s*(\d+)%',
    '깃털': r'(?:깃털|FEATHER)\s*(\d+)%',
    '우모': r'(?:우모)\(?([^\)]*)\)?', # Catch content in parentheses if available
    '거위': r'(?:거위)\s*(\d+)%',
    '오리': r'(?:오리)\s*(\d+)%',
    '알루미늄': r'(?:알루미늄|ALUMINUM)(?:\s*(\d+))?%', # Make percentage optional
    '천연가죽': r'(?:천연가죽)',
    '합성피혁': r'(?:합성피혁)',
    'I.P': r'(?:I.P)',
    'PHYLON': r'(?:PHYLON)',
    'PU': r'(?:PU)',
    'PVC': r'(?:PVC)',
    '두랄루민': r'(?:두랄루민|DURALUMIN)',
    'Carbon': r'(?:Carbon|카본)',
    'Carbonal': r'(?:Carbonal|카보날)',
    '신세틱': r'(?:신세틱|SYNTHETIC)',
    '러버': r'(?:러버|RUBBER)'
}

def extract_material_percentage(text, material_patterns):
    materials_info = {}
    if pd.isna(text) or text.strip() == '':
        return materials_info

    text = str(text).upper().replace(' ', '').replace(',', '') # Normalize text

    for material, pattern in material_patterns.items():
        match = re.search(pattern.upper(), text)
        if match:
            # For materials that often don't have explicit percentages, assume 100% presence if found.
            # '우모' also falls here as its content might be a ratio, not a single percentage
            if material in ['천연가죽', '합성피혁', 'I.P', 'PHYLON', 'PU', 'PVC', '두랄루민', 'Carbon', 'Carbonal', '신세틱', '러버', '우모']:
                materials_info[f'{material}_Percentage'] = 100
            else:
                try:
                    percentage_str = match.group(1)
                    if percentage_str is None or percentage_str == '':
                        materials_info[f'{material}_Percentage'] = 100 # Default to 100 if no percentage specified
                    else:
                        materials_info[f'{material}_Percentage'] = int(percentage_str)
                except (ValueError, IndexError):
                    materials_info[f'{material}_Percentage'] = 100 # Fallback for unexpected formats

    return materials_info
```

```python
material_data = df['Combined_Product_Material'].apply(lambda x: extract_material_percentage(x, material_patterns)).apply(pd.Series)
df = pd.concat([df, material_data], axis=1)

for material in common_materials:
    if f'{material}_Percentage' in df.columns:
        df[f'{material}_Percentage'] = df[f'{material}_Percentage'].fillna(0).astype(int)

df.drop(columns=['Product_Material', 'Main_Product_Material', 'Material'], inplace=True)
```

```python
# Rename the newly created material percentage columns to English
df.rename(columns={
    '신세틱_Percentage': 'Synthetic_Percentage',
    '합성피혁_Percentage': 'Synthetic_Leather_Percentage',
    '천연가죽_Percentage': 'Natural_Leather_Percentage',
    '우모_Percentage': 'Down_Feather_Percentage',
    '알루미늄_Percentage': 'Aluminum_Percentage',
    '두랄루민_Percentage': 'Duralumin_Percentage',
    'Carbonal_Percentage': 'Carbonal_Percentage',
    '러버_Percentage': 'Rubber_Percentage'
}, inplace=True)
```

```python
# --- 5. Standardize Manufacturing_Date ---
def standardize_manufacturing_date(date_str):
    if pd.isna(date_str) or str(date_str).strip() == '':
        return None

    date_str = str(date_str).strip().replace('년', '').replace('월', '').replace(' ', '')
    date_str = date_str.replace('22', '2022').replace('23', '2023').replace('24', '2024')

    try:
        return datetime.datetime.strptime(date_str, '%Y%m').strftime('%Y-%m')
    except ValueError:
        pass
    try:
        return datetime.datetime.strptime(date_str, '%Y-%m').strftime('%Y-%m')
    except ValueError:
        pass
    try:
        return datetime.datetime.strptime(date_str, '%Y %m').strftime('%Y-%m')
    except ValueError:
        pass
    return None
```

```python
df['Manufacturing_Date'] = df['Manufacturing_Date'].apply(standardize_manufacturing_date)
```

```python
# --- 6. Process and standardize Size information ---
df['Size'] = df['Size'].fillna('')
df['Size_Details'] = df['Size_Details'].fillna('')

df['Combined_Size'] = df['Size'] + ', ' + df['Size_Details']

def parse_sizes(size_str):
    if pd.isna(size_str) or size_str.strip() == '':
        return []
    size_str = size_str.replace(' ', '').replace(';', ',').replace('/', ',').replace('~', ',').replace('|', ',').replace('X', ',').replace('x', ',').strip()
    sizes = [s.strip() for s in size_str.split(',') if s.strip()]

    sizes = list(dict.fromkeys(sizes))
    sizes = [s for s in sizes if not re.match(r'(사이즈|중량|무게|프리|FREE):.*', s)]

    return sizes if sizes else None

df['Sizes_List'] = df['Combined_Size'].apply(parse_sizes)

df.drop(columns=['Size', 'Size_Details', 'Combined_Size'], inplace=True)

# --- 7. Extract keywords from Product_Features ---
feature_keywords = {
    'Waterproof': r'방수',
    'Breathable': r'투습',
    'Abrasion_Resistance': r'내마모성',
    'Elasticity': r'신축성',
    'Ultra_Lightweight': r'초경량',
    'Insulation': r'보온성',
    'Windproof': r'방풍',
    'Lightweight': r'경량',
    'Heating': r'발열'
}

for feature, pattern in feature_keywords.items():
    df[feature] = df['Product_Features'].apply(lambda x: bool(re.search(pattern, str(x))))

# --- 8. Process Selling_Price ---
df['Selling_Price'] = df['Selling_Price'].astype(int)

# --- 9. Extract Weight from Size_Weight ---
def extract_weight(text):
    if pd.isna(text) or str(text).strip() == '':
        return None
    weight_match = re.search(r'(?:중량|무게):\s*(\d+)\s*g|(\d+)\s*G', str(text), re.IGNORECASE)
    if weight_match:
        try:
            return int(weight_match.group(1) or weight_match.group(2))
        except ValueError:
            return None
    return None

df['Weight_g'] = df['Size_Weight'].apply(extract_weight)
df.drop(columns=['Size_Weight'], inplace=True)

# --- 10. Standardize Product Type (`Type` column) ---
type_mapping = {
    '기타모자': 'Hat',
    '기타장갑': 'Gloves',
    '캡 / 야구모자': 'Cap/Baseball Cap',
    '기타용품': 'Other_Accessories',
    '텐트': 'Tent',
    '기타캠핑용품': 'Other_Camping_Equipment',
    '등산스틱': 'Hiking_Pole',
    '슬링백 / 힙색': 'Sling_Bag/Hip_Sack',
    '버킷 / 사파리모자': 'Bucket/Safari_Hat',
    '백팩': 'Backpack',
    '크로스 / 숄더백': 'Cross/Shoulder_Bag',
    '파우치': 'Pouch',
    '토트백 / 보스턴백': 'Tote/Boston_Bag',
    '스카프 / 반다나': 'Scarf/Bandana',
    '우산 / 양산': 'Umbrella/Parasol',
    '등산배낭': 'Hiking_Backpack',
    'TK-117572': None,
    '60수 하이평직 BIO': None,
    '마스크': 'Mask'
}

df['Type'] = df['Type'].fillna('').apply(lambda x: type_mapping.get(x.strip(), None) if x.strip() != '' else None)

# --- Final Display (optional) ---
print("--- Final Preprocessed DataFrame Head ---")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\n--- Final Preprocessed DataFrame Info ---")
print(df.info())

print("\n--- Final Preprocessed DataFrame Columns ---")
print(df.columns.tolist())
```
